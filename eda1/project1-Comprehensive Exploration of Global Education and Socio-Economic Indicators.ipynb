```python
#histogram of latitude 
sns.histplot(data=df,x='Latitude ')
plt.title('Distribution of Latitude')
plt.xlabel('Latitude')
plt.ylabel('Frequency')
plt.show()
```

* it is positively skewed hisogram.

```python
# plotting a bargraph between countries(areas) and latitude.
plt.figure(figsize=(12, 8))
sns.barplot(x='Countries and areas', y='Latitude ', data=df.sort_values(by='Countries and areas', ascending=False).head(10))
plt.title('Top 10 Countries with the Highest Latitude')
plt.xlabel('countries and areas')
plt.ylabel('latitude')
plt.show()
```

* from the above bargraph 
* We come to know that the vatican  city is at the highest latitude where as venezuela is at lowest latitude.

```python

sns.histplot(data=df, x='OOSR_Pre0Primary_Age_Male')
plt.title('Histogram of Out-of-school Rate for Pre-Primary Age Males')
plt.xlabel('Out-of-school Rate')
plt.ylabel('Frequency')
plt.show()
```

* The above histogram is positively skewed it means that a majority of regions have lower out-of-school rates for pre-primary age males as shown in the above histogram.

```python
sns.histplot(data=df, x='OOSR_Primary_Age_Male')
plt.title('Histogram of Out-of-school Rate for Primary Age Males')
plt.xlabel('Out-of-school Rate')
plt.ylabel('Frequency')
plt.show()
```

The above histogram is positively skewed it means that a majority of regions have lower out-of-school rates for primary age males as shown in the above histogram

from above histograms we come to know that a majority of regions have lower out-of-school rates for both pre primary age males and primary age males as shown in the above histograms.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Explore the relationship between out-of-school rates and geographical location
plt.figure(figsize=(12, 8))

# Scatter plot with out-of-school rate for pre-primary age males vs. Latitude
sns.scatterplot(x='Latitude ', y='OOSR_Pre0Primary_Age_Male', data=df, hue='Countries and areas', palette='viridis'
plt.title('Relationship between Latitude and Out-of-school Rate for Pre-Primary Age Males')
plt.xlabel('Latitude')
plt.ylabel('Out-of-school Rate (Pre-Primary Age Males)')
plt.show()
```

* from the above scatterplot between Latitude and Out-of-school Rate for Pre-Primary Age Males
* The out of school rate is densed at lower latitude as compared to upper latitude

# Take away
The dataset that is  provided seems to contain information about various countries and areas, including their geographical coordinates and various education and socio-economic indicators. Here are some key takeaways based on the data:
* The dataset includes latitude and longitude coordinates for each country or area. Geographical locations range from Afghanistan, Albania, and Algeria to various other countries.

* The dataset includes indicators related to out-of-school rates (OOSR) for different age groups and genders. Completion rates for primary, lower secondary, and upper secondary education are provided for males and females.
* Literacy rates for youth aged 15-24 are provided for males and females.
* Birth rates, unemployment rates, and gross enrollment rates in primary and tertiary education are included.
* The dataset also provides information on the gross enrollment rate in primary education and tertiary education.
